```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,20)
```

```python
y=x*x
z=x+y
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

# Subplot

```python
plt.subplot(3,2,1)
plt.plot(x,x)
plt.subplot(3,2,2)
plt.plot(x,y)

```

# Object oriented plots 

```python
plt.plot(x,y)
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,y,x)
```

 # Multiple  Plots



# Figure Size

```python
fig=plt.figure(figsize=(5,3))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

# COLOR

```python
fig=plt.figure(figsize=(5,3))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='g',lw=15,linestyle='steps')
```

# Setting Limits

```python
fig=plt.figure(figsize=(5,3))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlim([0,6])
axes.set_ylim([0,60])
```
