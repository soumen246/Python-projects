# Categorial Plots 

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.pointplot(x='day',y='tip',data=tips)
```

```python
sns.swarmplot(x='smoker',y='tip',data=tips,hue='sex')
```

```python
sns.stripplot(x='smoker',y='tip',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex')
```
